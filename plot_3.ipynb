```python
import pandas as pd
```

```python
label = pd.read_csv("y_train.csv")
```

```python
X = pd.read_csv("X_train.csv")
```

```python
data = pd.concat([X, label], axis=1)
```

```python
import seaborn as sns
```

```python
sns.boxplot(x='label', y='num_prev_so', data=data)
```

```python
sns.scatterplot(x='num_prev_so', y='label', data=data)
```

```python
import matplotlib.pyplot as plt
```
