```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm = {"0": [76429, 32782, 25567], "1": [13611, 21500, 17030], "2": [6855, 12599, 15876]}
cm_df = pd.DataFrame(cm)
cm_df
```

```python
ax = sns.heatmap(cm_df, cmap="Blues", annot=True, annot_kws={"size": 16}, fmt="d")
ax.set_title(f"Confusion Matrix\n", size=20)
ax.xaxis.set_ticklabels(["Class0", "Class1", "Class2"], size=14)
ax.xaxis.set_ticks_position("top")
ax.yaxis.set_ticklabels(["Class0", "Class1", "Class2"], size=14, rotation=0)
plt.show()
```
